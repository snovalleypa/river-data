```python
df = pd.read_parquet('../data/svpa/svpa-33')
#df = pd.read_parquet('data/usgs/')
df.sort_index(inplace=True)
df
```

```python
import datetime
from datetime import timedelta

#identify crests
count = 0
previous_min = np.inf
previous_min_at = datetime.date(1920, 1, 1)
previous_max = 0
previous_max_at = datetime.date(1920, 1, 1)
max_row = None

#time gap
gap_hours = 6
max_delta = 1000
df_crests = []


for timestamp, row in df.iterrows():
    count += 1
    discharge = row['discharge']
#    print(timestamp, discharge)
    if discharge < previous_min:
        previous_min = discharge
        previous_min_at = timestamp
        continue
        
    if discharge > previous_max:
        previous_max = discharge
        previous_max_at = timestamp
        max_row = row
        continue
        
    if (timestamp - previous_max_at) < timedelta(hours=gap_hours):
        continue
        
    if previous_max - previous_min < max_delta:
        continue
        
    print('CREST:', previous_max_at, previous_max)
    df_crests.append(max_row)
    
    previous_min = discharge
    previous_min_at = timestamp
    previous_max = 0
    previous_mat_at = 0
    max_row = None
    
    if count > 5000:
        break
```
